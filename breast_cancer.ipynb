# Exploratory Analysis and Visualization

Here we are going to use Matplotlib and Seaborn to get the graphical representation of our data.
Matplotlib is basically for linear representation and we will use seaborn module above it for more advanced and clear representation of our data, which is a slightly large data frame. 



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.title('id')
plt.plot(new_frame.id)  
plt.xlabel('id')
plt.ylabel('length in 10000s');  #using ; in the end to remove matplotlib list of plots.
```

# Asking and Answering Questions

Trying to get some deep knowledge using some questions from data frame and trying to solve it using Pandas, Matplotlib and Seaborn



#### Q1: How many radius_mean are below 10?

```python
#passing a boolean expression to get the amount of persons having radius_mean less than 10.
less_tha_10 = new_frame[new_frame.radius_mean < 10] 
#now plotting a graph out of it
plt.plot(less_tha_10.radius_mean , 'or')
plt.title('radius_mean less than 10');
```

#### Q2: What is the difference between mean value of  perimeter_mean and texture_mean?

```python
mean_1 = new_frame.texture_mean.sum() /  len(new_frame.texture_mean)

mean_2 = new_frame.perimeter_mean.sum() / len(new_frame.perimeter_mean)

mean = mean_2 - mean_1

print('The difference between mean values is {}'.format(mean))  #using .format function to print in a single line
```

#### Q3: Find the number of person got the diagnosis in 'B' and 'M' category?

```python
#First getting the class distribution of diagnosis column and passing it in a variable name 'count'
count = new_frame['diagnosis'].value_counts()

#then plotting a histogram of the variable we defined earlier.
plt.title('Distribution of B and M')    #for the title
plt.hist(count , bins = 4);       #for the graph
```

#### Q4: Calculate the radius of worst area it has already spreads?

```python
#passsing a numerical expression to get the radius
radius = new_frame.area_worst / new_frame.perimeter_worst

#plotting the radius in a linear graph to see the range varies from (2,17)
plt.title('Variation of Radius')
plt.plot(radius);
```

#### Q5: Plot all the graphs in a grid

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
#plotting 1st graph
axes[0,0].plot(new_frame.id)
axes[0,0].legend('id')
axes[0,0].set_title('id length');


#plotting 2nd graph
axes[0,1].set_title('Radius')
axes[0,1].plot(radius);

#plotting 3rd graph
axes[0,2].set_title(" Mean ")
axes[0,2].hist([new_frame.concavity_mean, new_frame.compactness_mean, new_frame.smoothness_mean], 
         bins = 4,
         stacked = False)  

axes[0,2].legend(['concavity_mean', 'compactness_mean', 'smoothness_mean']);



#plotting 4th graph
axes[1,0].plot(concavity_all, 'or')
axes[1,0].legend(new_frame.diagnosis)
axes[1,0].set_title("Data about Concavity");


#plotting 5th graph
axes[1,1].plot(less_tha_10.radius_mean , 'or')
axes[1,1].set_title('radius_mean less than 10');




#plotting 6th graph
axes[1,2].set_title('Mean')
sns.scatterplot(x = new_frame.tandr_mean,  
                y = new_frame.radius_mean,
                hue=new_frame.diagnosis,
                s=150);
```
